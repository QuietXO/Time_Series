```python
# Split dataset into data + labels
X = concrete_raw.drop(columns=['Concrete compressive strength(MPa, megapascals) '])
y = concrete_raw[['Concrete compressive strength(MPa, megapascals) ']]

# Split dataset into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Normalization of dataset
normalN = concrete_raw['Concrete compressive strength(MPa, megapascals) '].max()

X_train = X_train.to_numpy()
y_train = np.divide(y_train.to_numpy(), normalN)

X_test = X_test.to_numpy()
y_test_normal = np.divide(y_test.to_numpy(), normalN)
y_test = y_test.to_numpy()
```

```python
# Turn data into Torch Format
X_train = torch.from_numpy(X_train).float()
y_train = torch.squeeze(torch.from_numpy(y_train).float())

X_test = torch.from_numpy(X_test).float()
y_test = torch.squeeze(torch.from_numpy(y_test).float())
y_test_normal = torch.squeeze(torch.from_numpy(y_test_normal).float())
```

<br>

### __*Create the Model*__

```python
class LinearRegression(nn.Module):
    def __init__(self, input_size, output_size):
        super(LinearRegression, self).__init__()
        self.fc1 = nn.Linear(input_size, 8)
        self.fc4 = nn.Linear(8, output_size)

    def forward(self, xx):
        xx = torch.sigmoid(self.fc1(xx))
        return self.fc4(xx)
```

```python
model = LinearRegression(8, 1)

criterion = nn.L1Loss()
optimizer = optim.Adam(model.parameters(), lr=0.00001)

loss_lst = list()
accuracy = list()

num_epochs = 2500
len_train = len(X_train)
split = 13

for epoch in range(num_epochs):
    total_loss = 0
    for delta in range(split):
        inputs = X_train[int(len_train/split)*delta:int(len_train/split)*delta+1]
        targets = y_train[int(len_train/split)*delta:int(len_train/split)*delta+1]

        # Forward pass
        outputs = model(inputs)
        loss = criterion(outputs, targets)

        # Backward and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if (epoch+1) % 50 == 0:
            total_loss += loss.item()

    if (epoch+1) % 50 == 0:
        loss_lst.append(total_loss/split)

        acc = criterion(model(X_test), y_test_normal)
        accuracy.append(acc.item())
        print(f"Epoch {epoch+1}/{num_epochs}| Loss: {total_loss/split}| Test Loss: {acc.item()}")

        total_loss = 0


fig_loss, axs_loss = plt.subplots(nrows=1, ncols=1, figsize=[15,5])
accuracy_plot, = axs_loss.plot(accuracy, label='Accuracy', color='#16b8f3')
loss_lst_plot, = axs_loss.plot(loss_lst, label='Loss', color='red')
axs_loss.grid(linestyle='--', linewidth=0.5)
axs_loss.legend(handles=[accuracy_plot, loss_lst_plot])

predictions = model(X_test)

print("Predictions:")
for i, x in enumerate(X_test):
    print(f"Input: {y_test[i]}, Predicted: {predictions[i][0]*normalN}")
```
